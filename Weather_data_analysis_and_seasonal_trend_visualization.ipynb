```python
import pandas as pd

# Load dataset
data = pd.read_csv("/content/sample_data/california_housing_train.csv")

# Check for missing values
print("Missing Values in Dataset:")
print(data.isnull().sum())

# Remove duplicate rows
data = data.drop_duplicates()

# Fill missing numerical values with mean
data.fillna(data.mean(numeric_only=True), inplace=True)

# Fill missing categorical values with mode
for column in data.select_dtypes(include='object'):
    data[column].fillna(data[column].mode()[0], inplace=True)

# Verify cleaned dataset
print("\nMissing Values After Cleaning:")
print(data.isnull().sum())

# Display cleaned data
print("\nCleaned Dataset:")
print(data.head())
```

```python
import pandas as pd

# Load dataset
data = pd.read_csv("/content/sample_data/california_housing_train.csv")

# Display first few rows
print("Dataset Preview:")
print(data.head())

# Display dataset information
print("\nDataset Information:")
print(data.info())

# Summary statistics
print("\nStatistical Summary:")
print(data.describe())

# Average values of housing parameters
print("\nAverage Median Income:", data['median_income'].mean())
print("Average Median House Value:", data['median_house_value'].mean())
print("Average Total Rooms:", data['total_rooms'].mean())
print("Average Population:", data['population'].mean())

# Group data by housing_median_age
housing_age_analysis = data.groupby('housing_median_age').mean(numeric_only=True)
print("\nAverage Housing Parameters by Median Age:")
print(housing_age_analysis)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/sample_data/california_housing_train.csv")

# Scatter plot for geographical distribution
plt.figure(figsize=(10, 8))
plt.scatter(data['longitude'], data['latitude'], alpha=0.4,
            c=data['median_house_value'], cmap='jet', s=data['population']/100) # Color by median house value, size by population

# Add labels and title
plt.title("California Housing: Longitude vs Latitude (Colored by Median House Value, Sized by Population)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Add a color bar for median house value
cbar = plt.colorbar()
cbar.set_label('Median House Value')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/sample_data/california_housing_train.csv")

# Group data by housing_median_age and calculate mean median_house_value
average_house_value_by_age = data.groupby('housing_median_age')['median_house_value'].mean()

# Create bar chart
plt.figure(figsize=(12, 6))
average_house_value_by_age.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title("Average Median House Value by Housing Median Age")
plt.xlabel("Housing Median Age")
plt.ylabel("Average Median House Value")
plt.xticks(rotation=90)

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/sample_data/california_housing_train.csv")

# Count number of records for each housing_median_age
housing_age_count = data['housing_median_age'].value_counts()

# Create pie chart
plt.figure(figsize=(8,8))
plt.pie(housing_age_count, labels=housing_age_count.index, autopct='%1.1f%%', startangle=90)

# Add title
plt.title("Distribution by Housing Median Age")

# Display chart
plt.show()
```
